```python
from sklearn import datasets
baza = datasets.load_iris()
```

```python
baza.keys()
```

```python
baza.target
```

```python
baza.target_names
```

```python
baza.feature_names
```

```python
X = baza.data
y = baza.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.5, random_state = 42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
modelKNN = KNeighborsClassifier(n_neighbors = 3)
modelKNN.fit(X_train, y_train)
```

```python
modelKNN.classes_
```

```python
modelKNN.effective_metric_
```

```python
modelKNN.score(X_test, y_test)
```

```python
y_predict_KNN = modelKNN.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predict_KNN)
```

```python
klasa = {0:'setosa', 1:'versicolor', 2:'virginica'}
```

```python
print('Za prvi cvijet iz skupa za testiranje predviđena je klasa {}, a stvarna klasa je {}.'.format(klasa[y_predict_KNN[0]], klasa[y_test[0]]))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_predict_KNN)
```

```python
from sklearn.metrics import precision_score, recall_score, average_precision_score
```

```python
precision_score(y_test, y_predict_KNN, average = None)
```

```python
recall_score(y_test, y_predict_KNN, average = None)
```

```python
from sklearn.metrics import f1_score
f1_score(y_test, y_predict_KNN, average = None)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predict_KNN, target_names = baza.target_names))
```
