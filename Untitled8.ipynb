```python
from sklearn import datasets
baza = datasets.fetch_openml('bodyfat')
```

```python
baza.keys()
```

```python
import numpy as np
podatci = np.array(baza.data)
podatci
```

```python
baza.feature_names
```

```python
tezina = podatci[:, 2] # zavisna variabla
tezina *= 0.46
```

```python
visina = podatci[:, 3] # nezavisna variabla
visina *= 2.54
```

```python
X = visina
y = tezina
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y)
plt.xlabel('Visina u cm')
plt.ylabel('Tezina u kg')
plt.xlim(160, 200)
plt.ylim(50, 130)
plt.show()
```

```python
X.shape
```

```python
X = X.reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X, y)
```

```python
reg.coef_ # nagib pravca
```

```python
reg.intercept_ # gdje siječe os x
```

```python
y_predviđeno = reg.predict(X)
```

```python
reg.score(X, y)
```

```python
visina_21 = X[20,0]
stvarna_tezina = y[20]
dobivena_tezina_21 = round(y_predviđeno[20],2)
```

```python
print("21. osoba u bazi je visoka {}cm, stvarna tezina je {}kg, a tezina dobivena regresijskim modelom je {}.".format(visina_21, stvarna_tezina, dobivena_tezina_21))
```

```python
residualno_odstupanje_21 = abs(stvarna_tezina - dobivena_tezina_21)
residualno_odstupanje_21
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y, y_predviđeno)
```
