```python
import numpy as np
podatci = np.random.randint(1, 500, (50, 2))
```

```python
podatci
```

```python
import matplotlib.pyplot as plt
plt.scatter(podatci[:, 0], podatci[:, 1])
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 2, random_state = 42, n_init = 1)
```

```python
kmeans.fit(podatci)
```

```python
grupa = kmeans.predict(podatci)
```

```python
grupa
```

```python
centroidi = kmeans.cluster_centers_
```

```python
centroidi
```

```python
plt.scatter(podatci[:, 0], podatci[:, 1], c = grupa, cmap = 'cool')
plt.scatter(centroidi[:, 0], centroidi[:, 1], marker = 'x', color = 'r', s = 200)
plt.show()
```

```python
kmeans.predict([[296, 103]])
```

```python
from sklearn import metrics
score = metrics.silhouette_score(podatci, grupa, metric = 'euclidean')
print('Silhouetter score: %.3f' % score)
```

```python
def vise_klustera(a, b, podatci):
  max_score = 0
  broj_klustera = 0
  scores_po_broju = []
  broj_klustera_po_broju = []

  for i in range (a, b + 1):
    kmeans = KMeans(n_clusters = i, random_state = 42, n_init = 1)
    grupa = kmeans.fit_predict(podatci)
    score = metrics.silhouette_score(podatci, grupa, metric = 'euclidean')
    print('Silhouetter score: %.3f' % score)

    scores_po_broju.append(score)
    broj_klustera_po_broju.append(i)

    if score > max_score:
      max_score = score
      broj_klustera = i

  print('Najveći score imamo sa {} klustera i iznosi {}'.format(broj_klustera, '%.3f' % max_score))
  plt.plot(broj_klustera_po_broju, scores_po_broju)
  plt.show()
```

```python
vise_klustera(2, 6, podatci)
```

```python
from yellowbrick.cluster import SilhouetteVisualizer
visualizer = SilhouetteVisualizer(kmeans, color = 'yellowbrick')
visualizer.fit(podatci)
visualizer.show()
```

```python
def vise_silueta(a, b, podatci):
  for i in range (a, b + 1):
    kmeans = KMeans(n_clusters = i, random_state = 42, n_init = 1)
    visualizer = SilhouetteVisualizer(kmeans, color = 'yellowbrick')
    visualizer.fit(podatci)
    visualizer.show()
```

```python
vise_silueta(2, 6, podatci)
```

```python
from yellowbrick.cluster import KElbowVisualizer
visualizer = KElbowVisualizer(kmeans, k = (2, 11), locate_elbow = True)
visualizer.fit(podatci)
visualizer.show()
```
