```python
import cv2
from google.colab.patches import cv2_imshow
slika = cv2.imread('/content/Cvijet')
v, s, d = slika.shape
print('Visina: {}, Širina: {}, Dubina: {}'.format(v, s, d))
cv2_imshow(slika)
(b, g, r) = slika[130, 130]
print("Pixel at (0, 0) - Blue: {}, Green: {}, Red: {}".format(b, g, r))
slika[130, 130] = (255, 255, 255)
cv2_imshow(slika)
(b, g, r) = slika[130, 130]
print("Pixel at (0, 0) - Blue: {}, Green: {}, Red: {}".format(b, g, r))
slika[100:120, 105:110] = (255, 0, 255)
cv2_imshow(slika)
slika2 = cv2.resize(slika, None, fx = 0.5, fy = 2)
gray = cv2.cvtColor(slika2, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```
